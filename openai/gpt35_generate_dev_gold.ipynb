```python
from concurrent.futures import ThreadPoolExecutor
import openai

def fetch_response(data_point):
    response = openai.ChatCompletion.create(
        model="ft:gpt-3.5-turbo-0613:personal:no-duplicates-test:7x2ZqXfU",
        messages=data_point['messages']
    )
    # print(response)
    sql_query_generated = response['choices'][0]['message']['content']
    return sql_query_generated


```

```python
with open('new-full-thing-no-new-lines', 'a') as f:

  for i in range(443,len(val_spider_data)):
  
    dp = val_spider_data[i]  
    sql_query = fetch_response(dp)
    
    f.write(sql_query.replace('\n', ' ') + '\n')
    print("i is: ", i)
```

```python
# Initialize an empty list to store the output SQL queries
output_sql_queries = []

# Create a ThreadPoolExecutor to run the OpenAI calls in parallel
with ThreadPoolExecutor(max_workers=50) as executor:
    # Submit the OpenAI calls to the executor and store the future objects in a list
    futures = [executor.submit(fetch_response, data_point) for data_point in val_spider_data]
    
    # Collect the responses as they complete, in the order they were submitted
    for future in futures:
        output_sql_queries.append(future.result())

# Now output_sql_queries contains the responses in the same order as val_spider_data
```

```python
len(output_sql_queries)
```

```python
# write output sql queries to text file:
# we'd need to open here in an append mode kind of thing
with open('old-output-of-ft:gpt-3.5-turbo-0613:personal:no-duplicates-test:7x2ZqXfU.txt', 'w') as f:
    for item in output_sql_queries:
        item = item.replace('\n', ' ')
        f.write("%s\n" % item)
```
